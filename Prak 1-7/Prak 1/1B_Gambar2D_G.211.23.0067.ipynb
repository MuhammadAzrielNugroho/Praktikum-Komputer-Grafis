```python
import numpy as np
import matplotlib.pyplot as plt

plt.grid(True)
plt.axis('on')

plt.axis([-10, 10, 10, -10])
plt.axes().set_aspect('equal')

#—————————————custom grid
x1 = -10
x2 = 10
y1 = 10
y2 = -10
dx = .5
dy = -.5

for x in np.arange(x1, x2, dx):
    for y in np.arange(y1, y2, dy):
        plt.scatter(x, y, s=1, color='lightgray')

#—————————————squer box
plt.plot([-5, 5], [-5, -5], linewidth=2, color='r')
plt.plot([5, 5], [-5, 5], linewidth=2, color='r')
plt.plot([5, -5], [5, 5], linewidth=2, color='r')
plt.plot([-5, -5], [5, -5], linewidth=2, color='r')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


plt.axis([0, 150, 100, 0])
r = 40
alpha1 = np.radians(0)
alpha2 = np.radians(360)
dalpha = np.radians(2)
xc = 75
yc = 50
plt.scatter(xc, yc, s=10, color='r')
for alpha in np.arange(alpha1, alpha2, dalpha):
    x = xc + r * np.cos(alpha)
    y = yc + r * np.sin(alpha)
    plt.scatter(x, y, s=5, color='r')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.axis([-10, 140, 90, -10])

plt.axis('on')
plt.grid(True)

plt.title('Contoh Sumbu')

#—————————————custom grid
dx = 5
dy = -5
for x in np.arange(-10, 140, dx):
    for y in np.arange(90, -10, dy):
        plt.scatter(x, y, s=1, color='lightgray')

#—————————————coordinat axes
plt.arrow(0, 0, 20, 0, head_length=4, head_width=3, color='b')
plt.arrow(0, 0, 0, 20, head_length=4, head_width=3, color='b')

plt.show()
```
